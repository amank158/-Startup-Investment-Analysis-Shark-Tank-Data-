```python
import pandas as pd
import numpy as np


file_path = "C:/Users/amank/OneDrive/Desktop/elevate labs intern/shark tank/Shark Tank India.csv"
data = pd.read_csv(file_path)


data.head()
```

```python

data.info()


data.isnull().sum()


data.describe(include='all')
```

```python

data.dropna(inplace=True)


```

```python

print("Duplicates found:", data.duplicated().sum())


data.drop_duplicates(inplace=True)
```

```python

data.columns = data.columns.str.strip()
```

```python

data['Industry'] = data['Industry'].str.lower().str.strip()


data['Startup Name'] = data['Startup Name'].str.lower().str.strip()
data['Business Description'] = data['Business Description'].str.lower().str.strip()
```

```python

print(data.columns)
```

```python
data.isnull().sum()
```

```python
data['Valuation Requested'] = pd.to_numeric(data['Valuation Requested'], errors='coerce')
data['Total Deal Amount'] = pd.to_numeric(data['Total Deal Amount'], errors='coerce')
data['Total Deal Equity'] = pd.to_numeric(data['Total Deal Equity'], errors='coerce')
data['Total Deal Debt'] = pd.to_numeric(data['Total Deal Debt'], errors='coerce')
```

```python
print("Unique Startups:", data['Startup Name'].nunique())
print("Unique Industries:", data['Industry'].nunique())

data['Industry'].value_counts().head(10)
```

```python
top_startups = data[['Startup Name', 'Total Deal Amount']].sort_values(by='Total Deal Amount', ascending=False).head(10)
top_startups
```

```python
shark_investments = {
    'Namita': data['Namita Investment Amount'].sum(),
    'Vineeta': data['Vineeta Investment Amount'].sum(),
    'Anupam': data['Anupam Investment Amount'].sum(),
    'Aman': data['Aman Investment Amount'].sum(),
    'Peyush': data['Peyush Investment Amount'].sum(),
    'Ritesh': data['Ritesh Investment Amount'].sum(),
    'Amit': data['Amit Investment Amount'].sum(),
    'Guest': data['Guest Investment Amount'].sum()
}

shark_investments
```

```python
industry_funding = data.groupby('Industry')['Total Deal Amount'].sum().sort_values(ascending=False)
industry_funding.head(10)
```
